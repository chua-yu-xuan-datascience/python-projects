Data Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\Jupyter Python stuff\Pandas- Data Visualisation\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#print(plt.style.available)
#plt.style.use('_classic_test_patch')
#can change style and colour of your visualisations
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
#line plot by default
# kind = 'line'
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked =True)
#stacked=True if you want stacked bar plots
#can only look at one column (flavor rating) if you want
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 200, c = 'green')
#s = size
#c = colour
```

```python
df.plot.hist(bins=10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y= "Flavor Rating", figsize = (8,8))
```
