Merge, Join, Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Admin\Downloads\Jupyter Python stuff\Pandas- Merging Data Frames\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\Admin\Downloads\Jupyter Python stuff\Pandas- Merging Data Frames\LOTR 2.csv")
df2
```

```python
df1.merge(df2)
#first one written is auto the df on the left and the one in the bracket is auto the df on the right
#and if never specify, it will do an inner join
# auto takes fellowshipif and firstname as the keys since they are unique and match
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
#has x and y because sys treat them differently since we didnt say they are matching keys, even though their values are the same
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID', how ='outer', lsuffix = '_Left', rsuffix = '_Right')
# join are more commonly used with indexes
# merge used with column names
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1, df2])
#df1 sitting right on top of df2, we are joining them together
```

```python
pd.concat([df1, df2], join = 'outer', axis=1)
#can still perform join on concat and merge them by specifying the axis1 as matching key
```
